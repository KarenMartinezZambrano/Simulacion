```python
import numpy as np
from scipy import integrate
import matplotlib.pyplot as plt
```

```python
def funcionA(x,f,k,P):
    integral=integrate.quad(funcionA2,0,P,args=(f,k,P))
    return 2*integral[0]/P

def funcionA2(x,f,k,P):
    return f(x)*np.cos(2*np.pi*k*x/(180*P))
                    
def funcionB(x,f,k,P):
    integral=integrate.quad(funcionB2,0,P,args=(f,k,P))
    return 2*integral[0]/P

def funcionB2(x,f,k,P):
    return f(x)*np.sin(2*np.pi*k*x/(180*P))       
                    
def FuncionSerieDeFourier(x,f,P,n): 
    a0=(2/P)*integrate.quad(f,0,P)
    a0=a0[0]/2
    for i in range (1,n+1):
        fA=funcionA(x,f,i,P)*np.cos(2*np.pi*i*x/(180*P))
        fB=funcionB(x,f,i,P)*np.sin(2*np.pi*i*x/(180*P))        
    return a0+fA+fB
```

```python
def primeraFuncion(x):
    return 1-x

def segundaFuncion(x):
    return x**2
```

```python
x = np.linspace(-2, 2, 100)
y = FuncionSerieDeFourier(x,primeraFuncion,1,6)
plt.plot(x, y)
plt.show()
```
